# **Single Layer Perceptron**

## **1. Importing Libararies**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
```

## **2. Reading Data and Header View**

```python
df = pd.read_csv("Student_performance_data _.csv")
df.head()
```

## **3. Shape of Data**

```python
df.shape
```

## **4. Null Values Distribution**

```python
df.isnull().sum()
```

## **5. Info about Data**

```python
df.info()
```

## **6. Description of Data**

```python
df.describe(include='all')
```

## **7. Train Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['StudentID', 'GradeClass']), df['GradeClass'],
                                                    test_size=0.2,
                                                    random_state=42)
```

## **8. Scaling**

```python
scale = MinMaxScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

## **9. Perceptron Model**

```python
perceptron = Perceptron()
perceptron.fit(X_train, y_train)
```

## **10. Prediction through Model**

```python
y_pred = perceptron.predict(X_test)
```

## **11. Classification Report**

```python
print(classification_report(y_test, y_pred))
```
